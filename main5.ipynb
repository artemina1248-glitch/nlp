```python
import spacy

# Load spaCy English model once globally (avoid reloading every call)
nlp = spacy.load("en_core_web_sm")

def extract_adj_noun_pairs(sentence):
    """
    Extract adjective-noun (amod) pairs from a sentence.
    Returns a list of tuples: [(adj, noun), ...]
    """
    doc = nlp(sentence)
    pairs = []

    for token in doc:
        # adjective modifying a noun (amod)
        if token.pos_ == "ADJ" and token.dep_ == "amod":
            head = token.head
            if head.pos_ == "NOUN":
                pairs.append((token.text, head.text))
    
    return pairs
```

```python
import pandas as pd
import spacy

# 1. Âä†ËΩΩ spaCy Ëã±ÊñáÊ®°Âûã
nlp = spacy.load("en_core_web_sm")

# 2. ÂÆö‰πâÊèêÂèñÂáΩÊï∞
def extract_adj_noun_pairs(sentence):
    """
    ÊèêÂèñÂΩ¢ÂÆπËØç-ÂêçËØç‰øÆÈ•∞ÂØπÔºàamodÔºâÔºåËøîÂõûÂàóË°® [("adj", "noun"), ...]
    """
    doc = nlp(str(sentence))  # ËΩ¨‰∏∫Â≠óÁ¨¶‰∏≤Èò≤Ê≠¢Áº∫Â§±ÂÄºÊä•Èîô
    pairs = []
    for token in doc:
        if token.pos_ == "ADJ" and token.dep_ == "amod":
            head = token.head
            if head.pos_ == "NOUN":
                pairs.append((token.text, head.text))
    return pairs

# 3. ÊâπÈáèÂ∫îÁî®Âà∞ df2['Article text']
df2['adj_noun_pairs'] = df2['Article text'].apply(extract_adj_noun_pairs)

# 4. ‰øùÂ≠òÁªìÊûúÂà∞Êñ∞Êñá‰ª∂Ôºà‰ªÖ‰øùÁïô Article text Âíå adj_noun_pairs ‰∏§ÂàóÔºâ
df2_all_pairs = df2[['Article text', 'adj_noun_pairs']]
df2_all_pairs.to_csv("df2_all_pairs.csv", index=False)

print("‚úÖ Â§ÑÁêÜÂÆåÊàêÔºåÊñá‰ª∂Â∑≤‰øùÂ≠ò‰∏∫ df2_all_pairs.csv")
```

```python
import pandas as pd

# ÂÅáËÆæ‰Ω†Â∑≤ÁªèÊúâ pair_freq
# pair_freq.columns = ['pair', 'count']

# Step 1: ÊèêÂèñ adj ÈÉ®ÂàÜÔºåÂπ∂Âä†Âà∞Êñ∞Âàó
pair_freq['adj'] = pair_freq['pair'].apply(lambda x: x[0])
  # ËΩ¨‰∏∫ tuple ÁÑ∂ÂêéÂèñÁ¨¨‰∏Ä‰∏™ÂÖÉÁ¥†

# Step 2: Êåâ adj ËÅöÂêà countÔºåÁªüËÆ°ÊØè‰∏™ adj ÁöÑÊÄªÈ¢ëÁéá
adj_freq = pair_freq.groupby('adj')['count'].sum().reset_index()

# Step 3: ËøáÊª§Âá∫Âá∫Áé∞Ê¨°Êï∞Â§ß‰∫é 50 ÁöÑ adj
adj_high_freq = adj_freq[adj_freq['count'] > 20].reset_index(drop=True)
high_freq_adjs = set(adj_high_freq['adj'])  # ËΩ¨‰∏∫ set Êñπ‰æøÂêéÁª≠Á≠õÈÄâ

# Step 4: Áî®È´òÈ¢ë adj Á≠õÈÄâ pair_freq ‰∏≠ÁöÑË°å
pair_freq_filtered = pair_freq[pair_freq['adj'].isin(high_freq_adjs)].reset_index(drop=True)

# ‚úÖ ËæìÂá∫ÁªìÊûú
print("üéØ È´òÈ¢ë adj Êï∞Èáè:", len(high_freq_adjs))
print("‚úÖ ‰øùÁïôÁöÑ pair Êï∞Èáè:", len(pair_freq_filtered))

# Step 5: ÂèØÈÄâÔºö‰øùÂ≠òÁªìÊûú
adj_high_freq.to_csv("adj_frequency_over_50_weighted.csv", index=False)
pair_freq_filtered.to_csv("pair_freq_filtered_by_adj.csv", index=False)
```

```python
from l2 import person_noun_list
person_nouns = set(person_noun_list)  # Áî® set Âä†ÈÄüÊü•Êâæ

pair_freq_filtered['noun'] = pair_freq_filtered['pair'].apply(lambda x: x[1])

# ÂàÜÁ±ª person / other
pair_freq_filtered['noun_category'] = pair_freq_filtered['noun'].apply(
    lambda n: 'person' if n in person_nouns else 'other'
)
```

```python
import pandas as pd

def compute_adj_person_ratios(df, label_col='noun_category', count_col='count'):
    """
    ÁªôÂÆöÂ∏¶Êúâ adj„ÄÅnoun„ÄÅnoun_categoryÔºàperson/otherÔºâ„ÄÅcount ÁöÑ pair_freq Ë°®Ôºå
    ËÆ°ÁÆóÊØè‰∏™ÂΩ¢ÂÆπËØç‰øÆÈ•∞ person ÁöÑÈ¢ëÁéá„ÄÅratio„ÄÅnormalized ratio„ÄÇ

    ËøîÂõû‰∏Ä‰∏™ÂåÖÂê´‰ª•‰∏ãÂàóÁöÑ DataFrame:
    ['adj', 'person_count', 'other_count', 'total', 'p_person_given_adj', 'normalized_ratio']
    """

    # Step 1: ÂÖ®Â±Ä P(person)
    total_counts = df[count_col].sum()
    person_counts = df[df[label_col] == 'person'][count_col].sum()
    p_person = person_counts / total_counts

    # Step 2: ÂàÜÁªÑÁªüËÆ°ÊØè‰∏™ adj ‰øÆÈ•∞ person / other ÁöÑÈ¢ëÁéáÔºàÂä†ÊùÉÔºâ
    pivot = df.pivot_table(index='adj', 
                           columns=label_col, 
                           values=count_col, 
                           aggfunc='sum', 
                           fill_value=0).reset_index()

    # Rename columns for consistency
    pivot = pivot.rename(columns={
        'person': 'person_count',
        'other': 'other_count'
    })

    # Step 3: ËÆ°ÁÆóÂêÑÈ°πÊØîÂÄº
    pivot['total'] = pivot['person_count'] + pivot['other_count']
    pivot['p_person_given_adj'] = pivot['person_count'] / pivot['total']
    pivot['normalized_ratio'] = pivot['p_person_given_adj'] / p_person

    # Step 4: Êåâ normalized ratio ÈôçÂ∫èÊéíÂ∫è
    pivot = pivot.sort_values('normalized_ratio', ascending=False).reset_index(drop=True)

    return pivot, p_person
```

```python

adj_ratios_df, global_person_ratio = compute_adj_person_ratios(pair_freq_filtered)

print(f"üåç Overall P(person): {global_person_ratio:.3f}")
print("\nüéØ Top 5 adj by normalized_ratio:\n")
print(adj_ratios_df.head())
```
